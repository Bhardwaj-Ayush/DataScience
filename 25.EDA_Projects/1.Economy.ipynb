```python
# importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# getting info about attributes
df.info()
```

```python
# converting price into int64
df['price'] = df['price'].replace(',','', regex =True).astype(int) #using regex cuz values are like 5,098
```

```python
df.info()
```

```python
# converting airline name into numeric data using one hot Encoding
df['airline'].unique()
```

```python
# One-Hot Encoding
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

airline = encoder.fit_transform(df[['airline']]).toarray()

encoded_df = pd.DataFrame(airline, columns = encoder.get_feature_names_out())
encoded_df.head()
```

```python
#concatinating the df with encoded dataframe
df1 = pd.concat([df,encoded_df], axis = 1)
df1.head()
```

```python
df1.drop('airline', inplace=True, axis = 1)  #now deleting the original column
```

```python
df1.head()
```

```python
# now handling date column
df1['day'] = df1['date'].str.split('-').str[0]
df1['month'] = df1['date'].str.split('-').str[1]
df1['year'] = df1['date'].str.split('-').str[2]
df1.columns
```

```python
df1.drop('date',inplace = True, axis = 1)
```

```python
df1.columns # the Date column is deleted
```

```python
# now handle ch_code and num_code
df1.drop(['ch_code','num_code'], inplace=True, axis = 1)
```

```python
df1.columns #both columns deleted
```

```python
df1.head()
```

```python
# now handle departure time and arrival time

df1['dep_hour'] = df1['dep_time'].str.split(':').str[0]
df1['dep_min'] = df1['dep_time'].str.split(':').str[1]
 
```

```python
# same with arr_time
df1['arr_hour'] = df1['dep_time'].str.split(':').str[0]
df1['arr_min'] = df1['dep_time'].str.split(':').str[1]
```

```python
# now droping the original column
df1.drop(columns = ['dep_time', 'arr_time'], inplace=True, axis = 1)
```
